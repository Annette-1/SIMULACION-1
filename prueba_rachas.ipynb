<span style="color:purple;">**Paso 1**</span> Determine la secuencia $S$ de $0$ y $1$ o (+,-)

Así
 Si $$ x_i≤x_{i-1}$$ escribir 0 (o -)

Si $$ x_i>x_{i-1}$$ escribir 1 (o -
+)


La secuencia $S$ contiene $n-1$ números

```python
# Función para generar la secuencia de 0 y 1
def generar_secuencia(datos):
    secuencia = np.zeros(len(datos) - 1, dtype=int)
    for i in range(1, len(datos)):
        if datos[i] > datos[i - 1]:
            secuencia[i - 1] = 1
        else:
            secuencia[i - 1] = 0
    return secuencia
```

<span style="color:purple;">**Paso 2**</span> Contar el número de rachas observadas $C_o$, una racha viene dada por el número de ceros y unos consecutivos que la forman

```python
# Función para contar el número de corridas
def contar_corridas(secuencia):
    corridas = 1  # La primera corrida siempre comienza en 1
    for i in range(1, len(secuencia)):
        if secuencia[i] != secuencia[i - 1]:  # Si cambia de 0 a 1 o viceversa, es una nueva corrida
            corridas += 1
    return corridas
```

Transponemos los datos para leer por columnas

```python
# Convertir a matriz de 20 filas x 5 columnas
matriz_original = np.array(data).reshape(20, 5)

# Leer por columnas usando ravel con order='F' (column-major)
datos_por_columnas = matriz_original.ravel(order='F')
```

```python
# Generar la secuencia binaria con los datos leídos por columnas
secuencia_binaria = generar_secuencia(datos_por_columnas)

secuencia_binaria= generar_secuencia(datos_por_columnas)
print("Secuencia binaria generada:", secuencia_binaria)
```

<span style="color:purple;">**Paso 3**</span> Calcular el valor esperado, y la varianza del número de rachas, así cómo el estadístico $z_0$ mediante:
$$μ_{Co}= \frac{2n-1}{3}$$
$$σ^2_{Co}=\frac{16n-29}{90}$$
 Y el estadístico de prueba es:
 $$Z_0=\frac{C_{o}-μ_{Co}}{σ^2_{Co}}$$

```python
# Calcular el número de corridas
num_corridas = contar_corridas(secuencia_binaria)

# Calcular el valor esperado y la varianza de las corridas
n = len(datos_por_columnas)
valor_esperado = (2 * n - 1) / 3
varianza = (16 * n - 29) / 90
desviacion = np.sqrt(varianza)

# Calcular el estadístico Z
Z = (num_corridas - valor_esperado) / desviacion
```

Verificamos la prueba de hipotesis dado el estadistico para prueba de rachas:
$$Z_0=\frac{C_{o}-μ_{Co}}{σ^2_{Co}}$$


Dado que $C_o$ sigue una distribución normal de media $μ_{c_o}$ y varianza $σ^2_{C_o}$ y $Z_0$ una distribución normal (0,1). El intervalo de aceptación:
$$
-≤Z_0\le Z_{\frac{α}{2}} $$
o bien
$$|Z_0| ≤Z_{\frac{α}{2}}$$

```python
# Prueba de hipótesis
alpha = 0.05
z_critico = norm.ppf(1 - alpha / 2)
intervalo_critico = [-z_critico, z_critico]
es_aleatorio = intervalo_critico[0] <= Z <= intervalo_critico[1]
```

```python
# Resultados
print("\n" + "=" * 60)
print("RESULTADOS")
print("=" * 60)
print("Número de corridas:", num_corridas)
print("Valor esperado de corridas:", valor_esperado)
print("Varianza:", varianza)
print("Desviación estándar:", desviacion)
print("Estadístico Z:", Z)
print("Intervalo crítico (α=0.05):", intervalo_critico)
print("¿La secuencia es aleatoria?:", es_aleatorio)
```

**Interpretación:**

El estadístico $Z$ calculado (-0.0827) cae dentro del intervalo crítico $[-1.96, 1.96]$, lo que indica que no hay evidencia suficiente para rechazar la hipótesis de aleatoriedad.
